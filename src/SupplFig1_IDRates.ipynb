```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import the data (can be downloaded on the main page of the Atlas)
rates = pd.read_csv('../data/SupplFig1_IDRates/SupplFig1_IDRates.tsv', sep = "\t", thousands=',')
rates["%Spectra ID'd"] = rates["%Spectra ID'd"].str.rstrip('%').astype('float')
rates = rates.sort_values(by = "%Spectra ID'd", ascending = False)
```

```python
# median
print(rates["%Spectra ID'd"].median())
```

```python
sns.boxplot(x="Dataset", y="%Spectra ID\'d", data=rates)
plt.xticks(rotation=90)
```

```python
sns.boxplot(x='Sample Category', y='%Spectra ID\'d', data=rates)
plt.xticks(rotation=90)
```

```python
# compare peptides between proteases
sns.boxplot(x='Protease', y='%Spectra ID\'d', data=rates)
```

```python
sns.boxplot(x='Protease', y='Distinct Peptides', data=rates)
# get median for both
rates.groupby('Protease')['Distinct Peptides'].median()
```

```python
sns.boxplot(x='Instrument Name', y='%Spectra ID\'d', data=rates)
plt.xticks(rotation=90)
```

```python
# lowest ID rate run?
lowest_id_run = rates.loc[rates['%Spectra ID\'d'].idxmin()]
print(f"Lowest ID rate run: {lowest_id_run['Dataset']} with {lowest_id_run['%Spectra ID\'d']}% ID'd")
```
